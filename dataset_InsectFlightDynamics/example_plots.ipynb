```python
# Obtain x and y coordinates of 1 fly

from get_coordinates import get_coordinates
import pandas as pd

# Load the CSV file into a DataFrame
file_path = 'Coordinate_second.csv'  # Replace with the actual file path
df = pd.read_csv(file_path)

# Extract all coordinates
type_name = 'Calocasiae' # 'Dmelanogaster'
movie_number = 1
fly_number = 5
x, y = get_coordinates(df, type_name, movie_number, fly_number)
```

```python
# Plot a few paths

from matplotlib import pyplot as plt

# Create a 2x3 grid of subplots (for 9 subplots)
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

# Flatten the grid to make indexing easier
axs = axs.flatten()

for movie_number in range(1,4):
    for fly_number in range(3):
        subplot_index = (movie_number-1) * 3 + fly_number
        x, y = get_coordinates(df, type_name, movie_number, fly_number+5)
        axs[subplot_index].plot(x, y, marker='o')
        axs[subplot_index].set_title(f"Plot {subplot_index}")

plt.tight_layout()
plt.show()
```
